```python
import pandas as pd
import seaborn as sns
from itertools import combinations
import matplotlib.pyplot as plt
```

```python
import os
path = '../data/'
data_list = os.listdir(path)
```

```python
def eda(data_name):
    print(f'----> Предварительный анализ для {data_name}')
    data = pd.read_csv(path + data_name)
    data = data.dropna()
    length = data[['lvsa', 'pvsa', 'result']].drop_duplicates().shape[0]
    general_length = data.shape[0]
    print(f'\n Набор уникальных триплетов ЛВСА, ПВСА, Результат длинной в {length}шт \n Общая длина датасета равна: {general_length}')
    print()
    print(data[['lvsa', 'pvsa', 'result']].drop_duplicates())
    print()
    print('Матрица корреляций и Диаграммы рассеивания по всем возможным комбинациям столбцов. Значения раскрашены по целевой переменной:')
    sns.heatmap(data.corr(), annot=True);
    arr_of_cols = data.drop(['result'], axis=1).columns.values
    
    comb = list(combinations(arr_of_cols, 2))
    fig, axs = plt.subplots(2,3, figsize=(15, 15))
    k = 0
    for i in range(2):
        for j in range(3):
            sns.scatterplot(x=data[comb[k][0]], y=data[comb[k][1]], hue=data.result, ax=axs[i,j]);
            k+=1
```

```python
eda(data_list[0])
```
